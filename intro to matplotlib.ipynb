# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plaotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # width and height

# 3. Plot the data
ax.plot(x,y)

# 4.Customize plot
ax.set(title = "Simple plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

# 5. Save & shwo the figure
fig.savefig("images/sample-plot.png")
```

### Two options for subplots

```python
# Subplot option 1 (following this)
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))

#plot to each diffrent axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplot Option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
# Plot to each different index
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib OO method)

* When plotting something quickly, okay to use the pyplot method
* When plotting something advance, use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);

# ax.set_xlim([45,100]);
```

```python
## OO method from scartch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);
# Customize
ax.set(title="Heart Diesease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target');

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
# Subplot of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,10),
                               nrows=2,
                               ncols=1,
                               sharex=True)
# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

# Customize ax0
ax0.set(title = "Heart Diesease and Cholesterol Levels",
        ylabel="Chelesterol")

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");



# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Diesease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

# Add a title to a figure
fig.suptitle("Heart Diesease Analysis", fontsize=16, fontweight="bold");
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

## OO method from scartch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter"); # changes colour scheme
# Customize
ax.set(title="Heart Diesease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target');

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
# Customize x n y axis limitations


# Subplot of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,10),
                               nrows=2,
                               ncols=1,
                               sharex=True)
# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter")

# Customize ax0
ax0.set(title = "Heart Diesease and Cholesterol Levels",
        ylabel="Chelesterol")

# Change x axis limits
ax0.set_xlim([50,80])

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");



# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"],
                       cmap="winter")

# Customize ax1
ax1.set(title="Heart Diesease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Change ax1 x axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60, 200])

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

# Add a title to a figure
fig.suptitle("Heart Diesease Analysis", fontsize=16, fontweight="bold");
```
